## Adaboost Algorithm
- Initalize all weights $w_{i}$ to $\frac{1}{n}$
- For $j$ in $m$ boosting rounds:
 - Train classifier, $C_{j}$
 - Predict class labels: $\hat{y} = C_{j}(X)$
 - Compute error rate: $\epsilon = w_{j} \cdot (\hat{y} == y)$
 - Compute j-th coefficent: $\alpha_{j} = \frac{1}{2} \text{ln} \frac{1 - \epsilon}{\epsilon}$
 - Update weights: $w_{j+1}= w_{j} \times \text{exp}\left (-\alpha_{j} \times \hat{y} \times y \right )$
 - Renormalize weights: $w = \frac{w}{\sum_{i} w_{i}}$
- Final model is $C_{m}$, whose prediction is: $\hat{y} = \left (\sum^{m}_{j=1} (\alpha_{j} \times C_{j}(X)) > 0 \right )$

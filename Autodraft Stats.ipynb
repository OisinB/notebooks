```python
import pandas as pd
%matplotlib inline
```

### Liberal Rules

```python
df = pd.read_csv('/Users/oisin-brogan/Downloads/user_recipes_liberal.csv')
```

```python
liberal_rules_results = df[df.classification_method == "seperate_cnn_lstm"]
```

```python
print ("Total suggestions made: %s" % liberal_rules_results.shape[0])
```

```python
liberal_rules_results.recipe_label.value_counts()
```

```python
(liberal_rules_results.recipe_label.value_counts()/liberal_rules_results.shape[0]).round(3)
```

```python
ax = liberal_rules_results.groupby('classified_at').recipe_id.count().plot(title = "Number of suggestions per day over time")
ax.set_xlabel("Date suggestion classified at")
```

```python
by_user = liberal_rules_results.groupby('user_id').recipe_label.value_counts()
by_user[by_user > 1].unstack().plot(kind = 'bar', figsize = [20,8], title = "Count of suggestions for users (with multiple suggestions)")
print("Number of users with multiple suggestions", by_user[by_user > 1].shape[0])
```

```python
print("Number of users with just one suggestion")
by_user[by_user <= 1].unstack().sum()
```

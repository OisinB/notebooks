```python
import imagehash
import boto3
from io import BytesIO
from PIL import Image


class ImageManager:

    """For handling images related to the recipe detection feature"""

    def __init__(self):
        pass

    def getS3Client(self):
        """Return a boto3 client for the s3 bucket containing user photos"""

        s3 = boto3.resource('s3', region_name='us-east-1',
                        aws_access_key_id='AKIAIONUF3ENWUFDLORQ',
                        aws_secret_access_key='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')

        return s3.Bucket('cookpad-global-private-images')

    def getS3Image(self, provider_id, photo_uid):
        """Return a bystream object of an image, downloaded from Amazon S3
        bucket containing user uploaded photos using provider and image id"""

        # We need generate the filename of the image in the S3 bucket
        s3_bucket = "production"
        s3_folder = 'photo_samples'
        provider_id = int(provider_id)
        source = '{3}/{0:03d}_{1}/{2}/{2}.jpg'.format(provider_id, s3_folder,
                                                      photo_uid, s3_bucket)

        # Create a boto s3 client and a bytesteam for storing the image data
        s3_client = self.getS3Client()
        data = BytesIO()

        # Download the image to memory
        s3_client.download_fileobj(source, data)

        # Reset the bytestream location to its start, and return it
        data.seek(0)
        return data

    def getImageHash(self, bytestream_image, hash_type='p'):
        """Use the image to generate hashes for the model"""

        pil_image = Image.open(bytestream_image)

        if hash_type == 'p':
            return str(imagehash.phash(pil_image))

        elif hash_type == 'w':
            return str(imagehash.whash(pil_image))
        
        elif hash_type == 'd':
            return str(imagehash.dhash(pil_image))

        elif hash_type == 'a':
            return str(imagehash.average_hash(pil_image))
        else:
            raise Exception("Attempted to generate hash using unknown type")
```

```python
from functools import reduce
```

```python
def is_suggestion_recipe(suggestion, user_recipes, max_extra = 1, min_percentage = .75):
    """suggestion is a list of image ids making a suggestion given by the rules
    user_recipes is a list of lists of image ids, detailing all the recipes
    manually identitified in the user's photos"""
    #Bools to show if we ever meet the condidtions individually, but not in
    #the same recipe
    best_extra_photos = False
    best_suff_cover = False

    flat_suggestion = reduce(lambda x,y: x+y, suggestion)
    singles = [x[0] for x in suggestion]

    for recipe in user_recipes:
        no_extra_photos = False
        suff_cover = False

        sf_matches = sum([i in recipe for i in flat_suggestion])
        ep_matches = sum([i in recipe for i in singles])
        extra_photos = len(suggestion) - ep_matches
        percentage = float(sf_matches) / len(recipe)
        if extra_photos <= max_extra:
            no_extra_photos = True
            best_extra_photos = True
        if percentage >= min_percentage:
            suff_cover = True
            best_suff_cover = True
        if no_extra_photos and suff_cover:
            #We matched to a labelled recipe - no need to check the others
            break

    return [best_extra_photos, best_suff_cover]
```

```python
def was_recipe_found(recipe, user_suggestions, max_extra = 1, min_percentage = .75):
    best_extra_photos = False
    best_suff_cover = False
    
    for suggestion in user_suggestions:
        flat_suggestion = reduce(lambda x,y: x+y, suggestion)
        singles = [x[0] for x in suggestion]
    
        no_extra_photos = False
        suff_cover = False

        sf_matches = sum([i in recipe for i in flat_suggestion])
        ep_matches = sum([i in recipe for i in singles])
        extra_photos = len(suggestion) - ep_matches
        percentage = float(sf_matches) / len(recipe)
        if extra_photos <= max_extra:
            no_extra_photos = True
            best_extra_photos = True
        if percentage >= min_percentage:
            suff_cover = True
            best_suff_cover = True
        if no_extra_photos and suff_cover:
            #We matched to a labelled recipe - no need to check the others
            break

    return [best_extra_photos, best_suff_cover]
```

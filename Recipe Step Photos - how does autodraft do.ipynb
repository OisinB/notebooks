```python
step_photos = pd.read_csv('/Users/oisin-brogan/Downloads/step_photos2/db.csv')

def read_exif_txt(txt_path):
    datetimes = {}
    with open(txt_path, 'r') as f:
        lines = f.readlines()
    for line in lines:
        if line.startswith('/'):
            key = line.split('//')[1][:-5] #getting just image id
            datetimes[key] = ""
        else:
            datetimes[key] = line.split('value ')[1].strip()
    return datetimes
    
def add_taken_at(db, txt_path):
    df = db.copy()
    dic = read_exif_txt(txt_path)
    series = pd.Series(dic)
    df = df.set_index('image_id')
    df.loc[:, 'taken_at'] = series
    df = df.reset_index()
    
    return df

step_photos = add_taken_at(step_photos, '/Users/oisin-brogan/Downloads/step_photos2/exif_data.txt')
step_photos.taken_at = step_photos.taken_at.map(rules.convert_exif_to_datetime)
```

```python
def overzealous_duplicates(recipe):
    test = rules.set_up_db_groups(recipe)
    test = rules.dup_removal_by_min_time(test, pd.Timedelta(seconds = 70))
    if test.empty:
        return False
    test = rules.dup_removal_by_hash_timed(test, 'recipe', 25, pd.Timedelta(minutes=150), 'phash')
    
    if test.shape[0] < 3:
        return True
    else:
        return False
```

```python
def mistake_kind(recipe_id, suggestions):
    if recipe_id not in suggestions.index:
        return "No EXIF data available for any photo in recipe"
    else:
        recipe = step_photos[step_photos.recipe_id == r_id]
        relevant_suggestions = suggestions.loc[r_id, 'suggestion_list']
        
        if len(relevant_suggestions) == 1:
            if sum(model_results.loc[r_id, 'suggestion_list'][0]) == 2:
                return "Correct suggestion"
        
        #Check we have at least three photos
        if recipe.shape[0] < 3:
            return "Less than 3 photos"
        
        #Check we aren't missing more than 25% of taken_at data
        missing_taken_ats = sum(pd.isnull(recipe.taken_at))
        if float(missing_taken_ats)/recipe.shape[0] > .25:
            return 'Missing over 25% of EXIF data'
        
        #we can drop nans now
        recipe = recipe[~pd.isnull(recipe.taken_at)]
        
        #Check total time isn't over 2.5 hours
        total_recipe_time = max(recipe.taken_at) - min(recipe.taken_at)
        if total_recipe_time > pd.Timedelta(minutes = 150):
            return "Recipe over time limit of 2.5 hours"
        
        if overzealous_duplicates(recipe):
            return "Duplicate removal removed too many photos"
        
        return 'Photo too dissimilar'
```
